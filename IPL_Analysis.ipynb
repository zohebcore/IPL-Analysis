### 7. Result probability after winning the Toss

```python
Toss_Win = matches[['toss_winner', 'winner', 'id']].groupby(['toss_winner', 'winner']).count().reset_index()
Toss_Win = Toss_Win.loc[Toss_Win['toss_winner'] == Toss_Win['winner']]
Toss_Win.columns = ['Team', 'Winner', 'Won']
Toss_Win.reset_index(inplace=True)
Toss_Win['Played'] = matches[['toss_winner', 'id']].groupby(['toss_winner']).count().reset_index()['id']
Toss_Win['Win_Prob'] = Toss_Win['Won'] / Toss_Win['Played']
Toss_Win.drop(['index', 'Winner', 'Won', 'Played'], axis=1, inplace=True)
Toss_Win.sort_values(by='Win_Prob', ascending=False, inplace=True)
sns.catplot(y='Team', x='Win_Prob', data=Toss_Win, kind='bar', orient='h', aspect=1.5).set(title='Result Probability after winning the Toss', xlabel='Winning Probability', ylabel='Teams')
plt.show()
```

### 8. Season wise performance of each Team.

```python
Season_Team = matches[['season', 'winner', 'id']]
Season_Team.columns = ['Season', 'Team', 'Wins']
Season_Team = Season_Team.groupby(by=['Season', 'Team']).count()
Season_Match1 = matches[['season', 'team1', 'id']]
Season_Match1.columns = ['Season', 'Team', 'Matches']
Season_Match2 = matches[['season', 'team2', 'id']]
Season_Match2.columns = ['Season', 'Team', 'Matches']
Season_Matches = pd.concat([Season_Match1, Season_Match2], axis=0)
Season_Matches = Season_Matches.groupby(by=['Season', 'Team']).count()
Season_Performance = pd.concat([Season_Matches, Season_Team], axis=1)
Season_Performance['Win_Rate'] = np.round(Season_Performance['Wins'] * 100 / Season_Performance['Matches'], decimals=2)
Season_Performance = Season_Performance.reset_index()
Season_Performance
```

```python
chart = sns.catplot(y='Win_Rate', col='Team', x='Season', data=Season_Performance, kind='bar', 
                    col_wrap=3, sharex=False, sharey=False)
chart.set_xticklabels(rotation=20).set(ylabel='Chance of Winning (%)', xlabel='Season')
plt.tight_layout()
plt.show()
```

# Merge both the Datasets

```python
ipl = pd.merge(left=matches, right=deliveries, how='outer', left_on='id', right_on='match_id')
ipl.head()
```

### 12. Average/Total runs per season

```python
Tot_runs = ipl[['season', 'total_runs']].groupby(by='season').sum().reset_index()
Tot_runs
sns.catplot(x='season', y='total_runs', data=Tot_runs, orient='v', kind='bar', aspect=2).set(title='Total Runs per Season', xlabel='Season', ylabel='Total Runs')
plt.show()
```

```python
match_season = matches[['season', 'id']].groupby(by='season').count().reset_index()
match_season
Avg_runs = pd.DataFrame(Tot_runs.total_runs/match_season.id)
Avg_runs.columns = ['Average_Runs']
Avg_runs['Season'] = match_season['season']
sns.catplot(x='Season', y='Average_Runs', data=Avg_runs, orient='v', kind='bar', aspect=2).set(title='Average Runs per Season', xlabel='Season', ylabel='Average Runs')
plt.show()
```

### 13. Run distribution per over.

```python
Over_Runs = ipl[['season', 'match_id', 'inning' ,'over', 'total_runs']].groupby(by=['season', 'match_id', 'inning','over']).sum().reset_index()

Run_Dist = pd.concat([Over_Runs[['over', 'total_runs']].groupby(by='over').mean(),
                      Over_Runs[['over', 'total_runs']].groupby(by='over').std()], 
                      axis=1).reset_index()
Run_Dist.columns=['Overs', 'Runs_Avg', 'Runs_Std']
Run_Dist
plt.errorbar(x='Overs', y='Runs_Avg', yerr='Runs_Std', data=Run_Dist)
plt.xlabel('Overs')
plt.ylabel('Runs')
plt.title('Run distribution per over')
plt.show()
```

### 14. Number of times, Target of 200 and plus by each team

```python
runs_inning = Over_Runs[['match_id', 'inning', 'total_runs']].groupby(by=['match_id', 'inning']).sum()
runs_200 = runs_inning[runs_inning.total_runs >= 200].reset_index()
runs_200 = pd.merge(left=runs_200[runs_200.inning == 1], right=runs_200[runs_200.inning == 2], on='match_id')
print("{} times a target of 200 and plus was scored by each team.".format(len(runs_200)))
```

### 15. Target of 200+ and how many teams and number of times chased.

```python
Over_Runs = ipl[['season', 'match_id', 'inning' ,'over', 'total_runs', 'winner']].groupby(by=['season', 'match_id', 'inning','over', 'winner']).sum().reset_index()
runs_inning = Over_Runs[['match_id', 'inning', 'total_runs', 'winner']].groupby(by=['match_id', 'inning', 'winner']).sum()
runs_200 = runs_inning[runs_inning.total_runs >= 200].reset_index()
runs_200 = pd.merge(left=runs_200[runs_200.inning == 1], right=runs_200[runs_200.inning == 2], on='match_id')
runs_200 = runs_200[runs_200.total_runs_y > runs_200.total_runs_x][['match_id', 'winner_x']].groupby(by='winner_x').count().reset_index()
runs_200.columns=['Teams', 'Chased_200']
runs_200.sort_values(by='Chased_200', ascending=False, inplace=True)
chart = sns.catplot(data=runs_200, x='Teams', y='Chased_200', kind='bar', aspect=2.5)
chart.set_xticklabels(rotation=20)
chart.set(title='Target of 200+ successfully chased by Teams', xlabel='Teams', ylabel='No. of Times successfully Chased')
```

### 16. Average Runs scored per over by each team.

### 17. Powerplay Analysis

```python
power_play = deliveries[deliveries.over <= 6]
print("Average Runs in Power Play: ", round(power_play.groupby(by=['match_id', 'inning'])['total_runs'].sum().mean(), 2))
print("Average Wickets in Power Play: ", round(power_play.groupby(by=['match_id', 'inning'])['player_dismissed'].count().mean(), 2))
power_play_score = power_play[['match_id', 'inning', 'batsman', 'batsman_runs', 'batting_team']].groupby(by=['match_id', 'inning', 'batsman', 'batting_team'])['batsman_runs'].sum().reset_index().groupby(by=['match_id', 'inning']).max().reset_index()
print("\nMax runs scored by a Batsman in Power play ->")
print(power_play_score[power_play_score.batsman_runs == power_play_score.batsman_runs.max()])

power_play_wkt = power_play[['match_id', 'inning', 'bowler', 'player_dismissed']].groupby(
    by=['match_id', 'inning', 'bowler'])['player_dismissed'].count().reset_index().groupby(
    by=['match_id', 'inning', 'bowler'])['player_dismissed'].max().reset_index()
print("\nMax Wickets taken by a Bowler in Power play ->")
print(power_play_wkt[power_play_wkt.player_dismissed == power_play_wkt.player_dismissed.max()])
```

### 22. Century, 50's and 30's run scored analysis per batsman.

```python
bats_run = deliveries[['match_id', 'batsman', 'batsman_runs']].groupby(by=['match_id', 'batsman']).sum().reset_index()
runs_100 = bats_run[bats_run['batsman_runs'] >= 100].groupby(by='batsman')['batsman_runs'].count().reset_index()
runs_100.columns = ['batsman', 'no100']
runs_50 = bats_run[bats_run['batsman_runs'] >= 50].groupby(by='batsman')['batsman_runs'].count().reset_index()
runs_50.columns = ['batsman', 'no50']
runs_30 = bats_run[bats_run['batsman_runs'] >= 30].groupby(by='batsman')['batsman_runs'].count().reset_index()
runs_30.columns = ['batsman', 'no30']
bats_score = pd.merge(left=pd.merge(left=runs_100, right=runs_50, how='outer', on='batsman'), 
                      right=runs_30, how='outer', on='batsman').replace(to_replace=np.nan, value=0)
bats_score[['no100', 'no50', 'no30']] = bats_score[['no100', 'no50', 'no30']].astype(int)
bats_score.sort_values(by=['no100', 'no50', 'no30'], ascending=False, inplace=True)
bats_score
```

```python
df = pd.melt(bats_score.head(10), id_vars="batsman", var_name="100/50/30", value_name="Count")
chart=sns.catplot(x='batsman', y='Count', hue='100/50/30', data=df, aspect=3, kind='bar')
chart.set_xticklabels(rotation=20)
chart.set(title='No. of 100s, 50s, 30s scored', ylabel='Count of 100/50/30', xlabel='Batsman')
plt.show()
```

### 23. Economy rate of Bowler.
